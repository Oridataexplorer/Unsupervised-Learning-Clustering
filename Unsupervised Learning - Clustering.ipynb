### Outlier

```python
LRFM=['L','R','F','M','C']
def dist_list(lst):
    plt.figure(figsize=[len(lst)*5,4])
    i = 1
    for col in lst:
        ax = plt.subplot(1,len(lst),i)
        sns.boxplot(dp_lrfmc_final[col])
        plt.title(col)
        i = i+1     
dist_list(LRFM)
```

```python
dari grafik distibusi diatas bisa disimpulkan data banyak outlier hingga distrubusi data nya positif skewness
```

```python
rfm_log=dp_lrfmc_final.copy()

cols=['R','F','M','C']
for col in cols:
    high_cut = rfm_log[col].quantile(q=0.99)
    low_cut=rfm_log[col].quantile(q=0.01)
    rfm_log.loc[rfm_log[col]>high_cut,col]=high_cut
    rfm_log.loc[rfm_log[col]<low_cut,col]=low_cut
    
    
LRFM=['L','R','F','M','C']
def dist_list(lst):
    plt.figure(figsize=[len(lst)*5,4])
    i = 1
    for col in lst:
        ax = plt.subplot(1,len(lst),i)
        sns.boxplot(rfm_log[col])
        plt.title(col)
        i = i+1
        
dist_list(LRFM)    
```

```python
rfm_log.shape
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler

ds_std=rfm_log
scale = MinMaxScaler()
kolom_all = [x for x in ds_std.columns]
for kolom in kolom_all:
  ds_std[kolom] = scale.fit_transform(np.array(ds_std[kolom]).reshape(-1,1))
```

# Modeling

## K MEANS & Clustering

```python
from sklearn.cluster import KMeans
ks = range(1,11)
inertias=[]
for k in ks :
    kc = KMeans(n_clusters=k,random_state=142)
    kc.fit(ds_std)
    cluster = kc.fit_predict(ds_std)
    inertias.append(kc.inertia_)
# Plot ks vs inertias
f, ax = plt.subplots(figsize=(10, 6))
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.style.use('ggplot')
plt.title('What is the Best Number for KMeans ?')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for n_clusters in range(2,6):
    clusterer = KMeans(n_clusters=n_clusters,init='k-means++',max_iter=300, random_state=142).fit(ds_std)
    preds = clusterer.predict(ds_std)

    # TODO: Calculate the mean silhouette coefficient for the number of clusters chosen
    score = silhouette_score(ds_std, preds, metric='mahalanobis')
    print ("For n_clusters = {}. The average silhouette_score is : {}".format(n_clusters, score))
```

```python
kc = KMeans(init='k-means++',n_clusters= 4, random_state=142)
kc.fit(ds_std)
cluster_labels = kc.labels_

data_c = ds_std.assign(K_Cluster = cluster_labels)
data_label=dp_lrfmc_final.assign(K_Cluster = cluster_labels)

data_f=data_c.groupby('K_Cluster').mean()
data_f['count']=pd.value_counts(data_c['K_Cluster'])

data_f
```
